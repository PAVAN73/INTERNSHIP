# # ICC Top 10 women’s ODI all-rounder along with the records of their team and rating

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder"
```

```python
page=requests.get(url)
page
```

```python
type(page)
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
player1=soup.find_all("div", class_="rankings-block__banner--name-large")
player1
```

```python
player2=soup.find_all("td", class_="table-body__cell rankings-table__name name")
player2
```

```python
players=player1+player2
players
```

```python


player3= []
for i in players:
    player3.append(i.text.strip())
player3
 
 
```

```python
player4=player3[0:10]
player4
```

```python
team1=soup.find_all("div", class_="rankings-block__banner--nationality")
team1
```

```python
team2=soup.find_all("td", class_="table-body__cell nationality-logo rankings-table__team")
team2
```

```python
team3=team1+team2
team3
```

```python
team4=team3[0:10]
team4
```

```python
teams= []
for i in team4:
    teams.append(i.text.strip())
teams
```

```python
rating1=soup.find_all("div", class_="rankings-block__banner--rating")
rating1
```

```python
rating2=soup.find_all("td", class_="table-body__cell rating")
rating2
```

```python
rating3=rating1+rating2
rating3
```

```python
rating4=rating3[0:10]
rating4
```

```python

ratings= []
for i in rating4:
    ratings.append(i.text.strip())
ratings
```

```python
career1=soup.find_all("div", class_="rankings-block__career-best")
career1
```

```python
career2=soup.find_all("td", class_="table-body__cell u-text-right u-hide-phablet")
career2
```

```python
career3=career1+career2
career3
```

```python
career4=career3[0:10]
career4
```

```python
career_best= []
for i in career4:
      career_best.append(i.text.strip())
career_best
 
```

```python
data =pd.DataFrame()
data['player'] = player4
data['team4']= teams
data['rating4']=ratings
data['career4']=career_best
data
```
