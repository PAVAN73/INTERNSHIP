# Top 10 ODI teams in women’s cricket along with the records for matches, points and rating

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.icc-cricket.com/rankings/womens/team-rankings/odi"
```

```python
page=requests.get(url)
page
```

```python
type(page)
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
type(soup)
```

```python
position1=soup.find_all("td", class_="rankings-block__banner--pos")
position1
```

```python
position2=soup.find_all("td", class_="table-body__cell table-body__cell--position u-text-right")
position2
```

```python

position=position1+position2
position
```

```python

positions= []
for i in position:
    positions.append(i.text.strip())
positions
```

```python
len(positions)
```

```python
team1=soup.find_all("span", class_="u-hide-phablet")
team1
```

```python
team2=team1[0:10]
team2
```

```python

teams_country= []
for i in team2:
    teams_country.append(i.text.strip())

teams_country
```

```python
len(team2)
```

```python
match1=soup.find_all("td", class_="table-body__cell u-center-text")
match1
```

```python
rating1=soup.find_all("td", class_="rankings-block__banner--rating u-text-right")
rating1
```

```python
rating2=soup.find_all("td", class_="table-body__cell u-text-right rating")
rating2
```

```python
rating=rating1+rating2
rating
```

```python

ratings= []
for i in rating:
    ratings.append(i.text.strip())
ratings
```

```python
len(ratings)
```

```python
data =pd.DataFrame()
data['position']= positions
data['team2'] = teams_country
data['match1']= match2
data['point1']=point2
data['rating']=ratings
data
```
