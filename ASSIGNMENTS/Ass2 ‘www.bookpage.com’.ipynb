# Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’



```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://bookpage.com/reviews"
```

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

```python
print(soup.prettify())
```

```python
book_names=soup.find_all("h4", class_="italic")
book_names
```

```python
book= []
for i in book_names:
    book.append(i.text.strip())
book


```

```python
len(book)
```

```python
author_name=soup.find_all("p", class_="sans bold")
author_name
```

```python
author= []
for i in author_name:
    author.append(i.text.strip())
author
```

```python
len(author)
```

```python
genre_name=soup.find_all("p", class_="genre-links hidden-phone")
genre_name
```

```python
genre= []
for i in genre_name:
    genre.append(i.text.strip())
genre
```

```python
len(genre)
```

```python
book_review=soup.find_all("div", class_="read-full")
book_review
```

```python
review= []
for i in book_review:
    review.append(i.text.strip())
review
```

```python
len(review)
```

```python
import pandas as pd
data=pd.DataFrame({})
data['book_names']=book
data['author name']=author
data["genre_name"]=genre
data["book_review"] =review 
data
```
