```python
page=requests.get("https://www.nobroker.in/property/sale/hyderabad/Mehdipatnam?searchParam=W3sibGF0IjoxNy4zOTU3ODQ3LCJsb24iOjc4LjQzMTE5NDQsInBsYWNlSWQiOiJDaElKalFEak5CeVh5enNSbHBkNG8wQURmUkEiLCJwbGFjZU5hbWUiOiJNZWhkaXBhdG5hbSJ9XQ==&radius=2.0")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

```python
print(soup.prettify())
```

```python
house=soup.find_all("h2", class_="heading-6 font-semi-bold nb__1AShY")
house
```

```python

house_titles= []
for i in house:
    house_titles.append(i.text.strip())
house_titles
```

```python
len(house_titles)
```

```python
area=soup.find_all("div", class_="nb__35Ol7")
area
```

```python

location= []
for i in area:
    location.append(i.text.strip())

location
```

```python
len(location)
```

```python
emi=soup.find_all("div",class_="font-semi-bold heading-6")
emi
```

```python
EMIS = []

for i in range(1,len(emi),3):

    EMIS.append(emi[i].get_text().replace("\n","").strip(' '))
```

```python
EMIS
```

```python
len(EMIS)
```

```python
price=soup.find_all("div", class_="nb__2NPHR")
price
```

```python
prices= []
for i in price:
    prices.append(i.text.strip())

prices
```

```python
prices = []

for i in range(3,len(price),4):

    prices.append(price[i].get_text().replace("\n","").strip(' '))
```

```python
prices
```

```python
len(prices)
```

```python
details =pd.DataFrame({})
details['house']= house_titles
details['area'] =location
details["emi"]=EMIS
details["price"]=prices
details
```
